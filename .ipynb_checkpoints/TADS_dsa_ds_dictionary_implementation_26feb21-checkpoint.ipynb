# Python dictionary implementation
---
**Goal:** implement a python dictionary from scratch

```python
class Dictionary:
    
    # create a list of empty slots 
    def __init__(self, size=16):
        self.size = size
        self.slots = [[] for _ in range(self.size)]
        
    # use python built-in hash() function to hash the key (alternative version: create custom hash function)
    # modulo the hash 
    def get_hash(self, key):
        return hash(key) % self.size
    
    # hash the key and return the slot corresponding to the hashed value
    def get_slot(self, key):
        return self.slots[get_hash(key)]
    
    # insert new (key:value) pair into dictionary (if key exists, update its value)
    def set_value(self, key, val):
        slot = get_slot(key)
        is_existing_key = False
        
        for idx, kv_pair in enumerate(slot):
            if kv_pair[0] == key:
                # replace the kv_pair with with updated (key, value) tuple
                slot[idx] = (key, value)
                is_existing_key = True
                break
                
        if not is_existing_key:
            slot.append(key, value)
            
    # extract value of existing key
    def get_value(self, key):
        slot = get_slot(key)
        
        for key_, value_ in slot:
            if key_ == key:
                return value_
            
        raise TypeError(f'Key: \'{key}\' not in dictionary.')
            
            
                
        
        
        
            
        
```
