```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import ipywidgets as wg
```

```python
width = 300
height = 100
```

```python
array = [1, 5, 7, 9, 10, 12, 13]
heights = [height] * len(array)
heights
```

```python
# import numpy as np
# import matplotlib.pyplot as plt
# import matplotlib.animation as animation
fig, ax = plt.subplots()
ax.scatter([1,2, 1.5], [2, 1, 1.5])
cir = plt.Circle((1.5, 1.5), 0.07, color='r',fill=False)
ax.set_aspect('equal', adjustable='datalim')
ax.add_patch(cir)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,8))
p = mpatches.Circle((2,height), 0.2, fill=True)
ax.scatter(array, heights, s = 100, c = 'r', alpha = 0.5)
ax.add_patch(p);
```
